```python
from functools import reduce
from itertools import count, islice, takewhile
from math import isqrt

def es_primo(n: int) -> bool:
    if n < 2: return False
    if n == 2: return True
    if n % 2 == 0: return False
    limite = isqrt(n)
    return reduce(lambda acc, d: acc and (n % d != 0), range(3, limite + 1, 2), True)
```

```python
def primos():
    for n in count(2):
        if es_primo(n):
            yield n
```

```python
# Probemos los primeros 15
print(list(islice(primos(), 15)))
```

```python
print(list(takewhile(lambda p: p < 50, primos())))
```

```python
sum(islice(primos(), 10))
```
